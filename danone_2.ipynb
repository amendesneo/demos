```python
neuvoo = pd.read_csv('data/Danone 2 - neuvoo.csv')
neuvoo.head()
```

```python
#neuvoo.groupby(['department_code', 'rome_code']).agg(['count'])
```

```python
neuvoo_results = neuvoo.groupby(['department_code', 'rome_code']).size().reset_index(name='counts')
neuvoo_results.head()
```

```python
def get_value(df, column_key, column_value, key):
    try:
        return df[df[column_key].str.startswith(key)][column_value].values[0]
    except Exception as e:
        print(e)
        return ''
    
get_value(departments, 'department', 'code', 'Ain')
```

```python
labone = pd.read_csv('data/Danone 2 - laboneboite.csv')
labone['department_code'] = [get_value(departments, 'department', 'code', department) for department in labone['department']]
labone.head()
```

```python
labone_results = labone.groupby(['department_code', 'rome_code']).size().reset_index(name='counts')
labone_results.head()
```

```python
def get_count(department, rome):
    #print(type(department))
    try:
        return neuvoo_results[(neuvoo_results['department_code'].astype(str).str.startswith(department)) & 
                   (neuvoo_results['rome_code'] == rome)]['counts'].values[0]

    except:
        return 0
get_count('6', 'D1509')
```

```python
def get_count_2(department, rome):
    #print(type(department))
    try:
        return labone_results[(labone_results['department_code'].astype(str).str.startswith(department)) & 
                   (labone_results['rome_code'] == rome)]['counts'].values[0]

    except:
        return 0
get_count_2('6', 'D1509')
```

```python
count = neuvoo_results[(neuvoo_results['department_code'] == 6) & 
               (neuvoo_results['rome_code'] == 'D1509')]['counts'].values[0]
count
```

```python
rome1 = 'M1706'
rome2 = 'D1509'

for department in count_matching['Departement']:
    
    #neuvoo
    count_matching.loc[count_matching['Departement'] == department, 'M1706_jobs'] = get_count(department, rome1)
    count_matching.loc[count_matching['Departement'] == department, 'D1509_jobs'] = get_count(department, rome2)
    
    #laboneboite
    count_matching.loc[count_matching['Departement'] == department, 'M1706_companies'] = get_count_2(department, rome1)
    count_matching.loc[count_matching['Departement'] == department, 'D1509_companies'] = get_count_2(department, rome2)
    

count_matching.to_clipboard(index=False)
count_matching
```
